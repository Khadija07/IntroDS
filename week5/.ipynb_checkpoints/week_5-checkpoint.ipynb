## Exercise 2 | Fairness-aware AI

### 2a)

This template generates data about the working hours and salaries of n=5000 people. The salary equals working hours times salary-by-hour, where salary-by-hour is given by 100 units plus normal distributed noise. If you run the template, it produces a scatter plot with gender=0 (men) in orange and gender=1 (women) in orange. The plot includes a trend line for each group, and an overall trend line for all data combined (in red). A linear regression model learned from the data *without* the protected characteristic (gender) should have slope close to 100.0

```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### Simulating the data

```python
#sample size n
n = 5000

# gender
gen = np.random.binomial(1, 0.5, size=n)

# work hours
hrs = np.random.binomial(60, 0.5, size=n)

# salary = hours * (100 + noise) (std.deviation 10)
sal = hrs * (100 + np.random.normal(0, 10, size=n))

# create a nice data frame
data = pd.DataFrame({"Gender": gen, "Hours": hrs, "Salary": sal})
```

### Scatterplot of the simulated data
Women (gender = 1) are shown with blue, men (gender = 0) are shown in orange.
Blue and orange lines are the trend lines of each group accordingly.
The overall trend line is shown in red.

```python
sns.regplot(x="Hours", y="Salary", data=data[data["Gender"]==1], color="darkBlue", scatter_kws={'alpha':0.5})

sns.regplot(x="Hours", y="Salary", data=data[data["Gender"]==0], color="darkOrange", scatter_kws={'alpha':0.5})

sns.regplot(x="Hours", y="Salary", data=data, marker="None", color="red")

plt.show()
```

### Linear regression
Fit the overall regression model, which is what an algorithm with no access to the gender ("protected characteristic") would learn from the data.

```python
reg = LinearRegression().fit(hrs.reshape(-1,1), sal.reshape(-1,1))

# print out the slope: it should be close to 100.0 without learning the 'protected characteristic' (gender)
print("slope: %.1f" % reg.coef_)
```

### Three scenarios

Now edit the code to simulate the following scenarios:

i. the salary of women is reduced by 200 euros ("direct discrimination")

ii. the working hours of men are binomially distributed with parameters (60, 0.55) while the working hours of women are binomially distributed with parameters (60, 0.45) ("no discrimination")

iii. both of the above changes at the same time ("indirect discrimination")

You should be able to demonstrate that the slope of the linear regression model is only changed in one of these scenarios.

```python
# Sample size n
n = 5000

# Gender
gen = np.random.binomial(1, 0.5, size=n)

# Salary Reduction for Women (Direct Discrimination)
# Calculate working hours for men and women without discrimination
hrs = np.random.binomial(60, 0.5, size=n)

# Calculate salary for men and women
sal_men = hrs * (100 + np.random.normal(0, 10, size=n))
sal_women_direct_discrimination = (hrs * (100 + np.random.normal(0, 10, size=n))) - 200

# Combine salaries based on gender for direct discrimination scenario
salaries_direct_discrimination = np.where(gen == 1, sal_women_direct_discrimination, sal_men)

# Create a DataFrame for direct discrimination scenario
data_direct_discrimination = pd.DataFrame({"Gender": gen, "Hours": hrs, "Salary": salaries_direct_discrimination})


# Create plots and fit linear regression models for all scenarios
def plot_and_fit(data, title):
    sns.regplot(x="Hours", y="Salary", data=data[data["Gender"] == 0], color="darkOrange", scatter_kws={'alpha': 0.5})
    sns.regplot(x="Hours", y="Salary", data=data[data["Gender"] == 1], color="darkBlue", scatter_kws={'alpha': 0.5})

    X = data[["Hours"]]
    y = data["Salary"]
    model = LinearRegression()
    model.fit(X, y)

    plt.plot(X, model.predict(X), color="red", label="Overall Trend Line")
    plt.title(title)
    plt.legend()
    plt.show()

    print("Overall trend line slope:", model.coef_[0])

# Scenario 1
plot_and_fit(data_direct_discrimination, "Direct Discrimination")
```

```python
# Working Hour Distribution Difference (No Discrimination)
# Calculate working hours for men and women with different distributions but no discrimination
hrs_men_no_discrimination = np.random.binomial(60, 0.55, size=n)
hrs_women_no_discrimination = np.random.binomial(60, 0.45, size=n)

# Calculate salary for men and women without discrimination
sal_men_no_discrimination = hrs_men_no_discrimination * (100 + np.random.normal(0, 10, size=n))
sal_women_no_discrimination = hrs_women_no_discrimination * (100 + np.random.normal(0, 10, size=n))

# Combine salaries based on gender for no discrimination scenario
salaries_no_discrimination = np.where(gen == 1, sal_women_no_discrimination, sal_men_no_discrimination)

# Create a DataFrame for no discrimination scenario
data_no_discrimination = pd.DataFrame({"Gender": gen, "Hours": np.where(gen == 1, hrs_women_no_discrimination, hrs_men_no_discrimination), "Salary": salaries_no_discrimination})

# Scenario 2
plot_and_fit(data_no_discrimination, "No Discrimination")
```

```python

# Both Changes (Indirect Discrimination)
# Calculate working hours for men and women with different distributions for indirect discrimination
hrs_men_indirect_discrimination = np.random.binomial(60, 0.55, size=n)
hrs_women_indirect_discrimination = np.random.binomial(60, 0.45, size=n)

# Calculate salary for men and women with salary reduction for women for indirect discrimination
sal_men_indirect_discrimination = hrs_men_indirect_discrimination * (100 + np.random.normal(0, 10, size=n))
sal_women_indirect_discrimination = (hrs_women_indirect_discrimination * (100 + np.random.normal(0, 10, size=n))) - 200

# Combine salaries based on gender for indirect discrimination scenario
salaries_indirect_discrimination = np.where(gen == 1, sal_women_indirect_discrimination, sal_men_indirect_discrimination)

# Create a DataFrame for indirect discrimination scenario
data_indirect_discrimination = pd.DataFrame({"Gender": gen, "Hours": np.where(gen == 1, hrs_women_indirect_discrimination, hrs_men_indirect_discrimination), "Salary": salaries_indirect_discrimination})

# Scenario 3
plot_and_fit(data_indirect_discrimination, "Indirect Discrimination")
```
